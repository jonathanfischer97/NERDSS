```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
ka1 = 0.1
kb1 = 33
ka2 = 0.1362
kb2 = 48.17
L = 100

Kd1 = (kb1/10**6)/ka1
Kd2 = (kb2/10**6)/(ka2/2)


T = np.array(range(1000))

Kd = (Kd1*Kd2)/(Kd2 + (T/L**2))
```

```python
fig = plt.figure(figsize=(8,4),dpi=200)
ax = fig.add_subplot(111)

ax.plot(T, Kd)
#ax.plot(AP,c='cyan',alpha=0.5,lw=0.5,ls='--', label = 'AP');
```
